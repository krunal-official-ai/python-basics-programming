# Return Multiple Values (Tuple Unpacking)

```python
def hello():
    a = 3
    b = 5
    return a+b

Hi = hello()
print(Hi)
```

```python
def hello():
    a = 3
    b = 5
    return a
    return b

Hi = hello()
print(Hi)
```

```python
def hello():
    a = 3
    b = 5
    return a,b

Hi = hello()

print(Hi)
```

```python
def hello():
    a = 3
    b = 5
    #return multiple values
    return a,b  #only This is possible in python.



Hi_1,Hi_2 = hello() #tuple unpacking
print(Hi_1)
print(Hi_2)
```

```python
n = int(input("Enter a number: "))
fact = 1

for i in range(1, n + 1):
    fact = fact * i

print("Factorial of", n, "is", fact)
```

# lambda | Anonymous function => function without name

# A lambda function is a small, one-line function without a name.

# ✔ It is temporary
# ✔ Used for short tasks
# ✔ Works in one line only
# ✔ No def required

```python
def cube(n):
    return n**3

cube(5)
```

# Syntex: lambda arguments : expression

>> we can keep multiple arguments
>> but Expression must be only one

```python
ans = lambda n : n**3
print(ans(3))
```

```python
addition = lambda a,b : a+b
print(addition(3,5))
```

```python
def getme(n): #n=4
    return lambda a: a*n

A = getme(4)
print (A(2))

B = getme(6)
print (B(3))
```

# map() takes a function and applies it to every element of an iterable (list, tuple, etc.)
# It does this without writing loops manually.

```python
map(function, iterable)
```

```python
# With map()
nums = [1, 2, 3, 4]
result = list(map(lambda x: x * 2, nums))
print(result)
```

# filter() is used to select items from a sequence based on True/False condition.

```python
filter(function, iterable)
✔ function → returns True or False
✔ iterable → list, tuple, set, generator, etc.
```

```python
# With filter()
nums = [1, 2, 3, 4, 5, 6]
even = list(filter(lambda x: x % 2 == 0, nums))
print(even)
```

# sorted() returns a new sorted list from any iterable.

```python
# With sorted()
students = [(1, "Raj"), (3, "Amit"), (2, "John")]

sorted_list = sorted(students, key=lambda x: x[0])
print(sorted_list)
```
