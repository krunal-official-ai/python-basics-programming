# ✅ Calculator using UDF and Ladder If-Else

```python
# User-Defined Function for Calculator

def calculator(a, b, operator):

    if operator == '+':
        result = a + b
    elif operator == '-':
        result = a - b
    elif operator == '*':
        result = a * b
    elif operator == '/':
        if b != 0:
            result = a / b
        else:
            result = "Error! Division by zero."
    else:
        result = "Invalid operator"

    return



# Input from user
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))
op = input("Enter operator (+, -, *, /): ")


# Function call
output = calculator(num1, num2, op)
print("Result:", output)
```

# Types of function Arguments
# 1. *args
# 2. **kwargs

```python
def Display(a, b, c):

    print(a,b,c)

Display(5, 2.3, True)
```

```python
def Display(a, b, c):

    print(a,b,c)

Display(5, 2.3, True, 15)
```

```python
def Display(*n):
    print(n)

Display(5, 2.3, True, 15)
```

```python
def Display(*args):  # * is the unpacking operator
    print(args)

Display(5, 2.3, True, 15)
print(type(Display))
```

```python
def Display(*args):

    for i in args:
        print(i)

Display(5, 2.3, True, 15)
print(type(Display))
```

```python
def Display(**n):
    print(n)

Display(no= 1, name='krunal', age=20)
```

```python
def Display(**kwargs):
    print(kwargs)

Display(no= 1, name='krunal', age=20)
```

```python
def Display(**kwargs):
    print(kwargs.keys())

Display(no= 1, name='krunal', age=20)
```

```python
def Display(**kwargs):
    print(kwargs.values())

Display(no= 1, name='krunal', age=20)
```

```python
def Display(*args, **kwargs):
    print(args)
    print(kwargs)

Display(1, 2, 3, 4, 5, no= 1, name='krunal', age=20)
```

```python
def Display(*args, **kwargs):
    print(args)
    print(kwargs)

Display(1, 2, 3, 4, 5, no= 1, name='krunal', age=20, 6, 7, 8, 9, 10)
```

# Docstring => Document String

```python
def Display(*args, **kwargs):
    """
    This is simple argument & keyword argument function.

    """
    print(args)
    print(kwargs)

Display(1, 2, 3, 4, 5, no= 1, name='krunal', age=20)
```

```python
def Display(*args, **kwargs):
    """
    This is simple argument & keyword argument function.

    """
    print(args)
    print(kwargs)

Display(1, 2, 3, 4, 5, no= 1, name='krunal', age=20)

print(Display.__doc__)
```

# 🌍 Global Keyword

```python
#global area
a = 5

def Jarvis():

    #local area
    print(a)

Jarvis()
```

```python
#global area
a = 5
def jarvis():
    a = 6
    print(a)
print(a)
jarvis()
```

```python
# New Variable in Local

#global area
a = 5
b = 7
def Jarvis():

    a = 8
    #local area
    print("Local a :",a)

print("Global a :",a)
print("Global b :",b)
Jarvis()
```

```python
#global area
a = 5

def Jarvis():

    a = 8
    #local area
    print("Local  a :",a)

print("Global a :",a)
Jarvis()
print("Global a :",a)
```

```python
#global area
a = 5

def Jarvis():

    global a
    a = 8
    #local area
    print("Local  a :",a)

print("Global a :",a)
Jarvis()
print("Global a :",a)
```

```python
#global area
a = 5
b = 10

def Jarvis():

    global a,b
    a = 8
    b = 2

    #local area
    print("Local  a :",a)
    print("Local  b :",b)
    print("            ")



print("previous Global a :",a)
print("previous Global b :",b)
print("            ")


Jarvis()

print("Now Global a :",a)
print("Now Global b :",b)
```
