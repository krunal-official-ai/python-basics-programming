# 🔒 Encapsulation in Python

# # Encapsulation is one of the core concepts in Object-Oriented Programming (OOP).

# ➵ Binding data (variables) and methods (functions) together in one unit (class). You can hide internal details using private variables.

# ➵ Encapsulation means hiding the internal details of an object and only allowing access through methods.



# ✅ Purpose of Encapsulation


# ➵ Hide internal object details

# ➵ Protect object state from unintended or harmful changes

# ➵ Provide controlled access through methods (getters/setters)

# 🎯 Simple Definition:

```python
Encapsulation = Data + Code together in a class + Restricted Access
```

# 👨‍💻 How Encapsulation Works

```python
# Prefix    Access Level    Meaning
  name      Public          Accessible from anywhere
```

```python
# Prefix    Access Level               Meaning
  _name     Protected (convention)     Should not be accessed outside class
```

```python
# Prefix    Access Level     Meaning
  __name    Private          Name mangled; not directly accessible
```

# 👨‍💻 CODE

```python
class Student:
    def __init__(self, name, age):
        self.name = name        # public
        self.__age = age        # private (using double underscore)

    # Getter method (to read private data)
    def get_age(self):
        return self.__age

    # Setter method (to update private data)
    def set_age(self, age):
        if age > 10:
            self.__age = age
        else:
            print("Invalid age")

# Create object
s1 = Student("krunal", 20)

print(s1.name)          # ✅ Public - can access
print(s1.get_age())     # ✅ Access private data using method

s1.set_age(11)          # ✅ Update private data safely
print(s1.get_age())

print(s1.__age)       # ❌ Error! Cannot access private data directly
```

# ✅ CODE :
# ✅ Bank Account

# You can deposit, withdraw, and check balance — but you can't directly change the balance from outside. That prevents errors like setting it to a negative number.

```python
class BankAccount:
    def __init__(self):
        self.__balance = 0  # private

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount

    def get_balance(self):
        return self.__balance

# Usage
acc = BankAccount()
acc.deposit(100)
print(acc.get_balance())   # Output: 100
```

# ✅ CODE :
# ✅ Student Info

```python
class Student:
    def __init__(self, name):
        self.__name = name  # private

    def get_name(self):
        return self.__name

    def set_name(self, name):
        self.__name = name

# Usage
s = Student("KARMA")
print(s.get_name())   # Output: KARMA
s.set_name("KRUNAL")
print(s.get_name())   # Output: KRUNAL
```

# ✅ CODE :
# ✅ Online Shopping System

```python
class Order:
    def __init__(self, order_id, customer_name):
        self.__order_id = order_id         # private attribute
        self.__customer_name = customer_name
        self.__items = []

    def add_item(self, item):
        self.__items.append(item)

    def get_order_details(self):
        return {
                "order_id": self.__order_id,
                "customer_name": self.__customer_name,
                "items": self.__items
               }

OS = Order(100,"krunal")
print(OS.get_order_details())
print(OS.add_item("magiee"))
print(OS.get_order_details())
```

# The internal list of items is protected, and can only be changed through methods like add_item().



# ✅ Conclusion :

# ➵ __age is private (not directly accessible).

# ➵ We use getter (get_age) and setter (set_age) methods to control access.

# ➵ Variables like __balance, __name, __celsius are private (using double underscore).

# ➵ This is encapsulation.
