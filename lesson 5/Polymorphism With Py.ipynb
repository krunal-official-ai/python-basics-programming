# Polymorphism means "many forms".

# Polymorphism in Python is an object-oriented programming concept that allows objects of different classes to be treated as if they were objects of the same class, through a common interface.

# It comes from the Greek words "poly" (many) and "morph" (form), meaning many forms.

# Let's dive into method overloading and method overriding in Python, two key concepts of polymorphism in object-oriented programming

# **🔁 Types of Polymorphism in Python:**

# 1. Compile-time Polymorphism
#    # Method Overloading ❌    
#    # Operator Overloading ✅  
# 2. Run Time polimorphm
#    # Method Overriding ✅

# 🔁**method Overriding (Run-Time polimorphs)**

# **A child class replaces a method from the parent class with its own version of that method.**

# **Method overriding in Python occurs when a subclass (child class) provides a specific implementation of a method that is already defined in its superclass (parent class). It’s a form of runtime polymorphism.**

# Vehicle → Car

```python
class Vehicle:
    def start(self):
        print("Starting a generic vehicle...")

class Car(Vehicle):
    def start(self):
        print("Starting a car with keyless ignition.")


v = Vehicle()
v.start()  # Output: Starting a generic vehicle...


c = Car()
c.start()  # Output: Starting a car with keyless ignition.
```

# BankAccount → SavingsAccount

```python
class BankAccount:
    def interest_rate(self):
        print("Base interest rate is 2%")

class SavingsAccount(BankAccount):
    def interest_rate(self):  # Overriding
        print("Savings account interest rate is 5%")


acc = BankAccount()
acc.interest_rate()  # Output: Base interest rate is 2%

sacc = SavingsAccount()
sacc.interest_rate()  # Output: Savings account interest rate is 5%
```

```python
class principal:
    def lacture(self):
        print("Today's class is Robotics")

class Krunal(principal): # child Class
    def lacture(self):   # Overriding the 'lacture' method
        print("absent")

class Karma(principal):  # child Class
    def lacture(self):   # Overriding the 'lacture' method
        print("present")

# Usage
a = principal()
d = Krunal()
c = Karma()

a.lacture()
d.lacture()
c.lacture()
```

# Employee → Manager

```python
class Employee:
    def work(self):
        print("Employee is working 9 to 5.")

class Manager(Employee):
    def work(self):
        super().work()  # Call the parent method
        print("Manager is attending meetings.")

m = Manager()
m.work()
```

# 🔎 Manager uses the base functionality of Employee and adds more responsibility.

# ✅CODE : Overriding with Constructor (_ _init_ _)

```python
class Person:
    def __init__(self, name):
        self.name = name

    def show(self):
        print(f"Name: {self.name}")


class Student(Person):
    def __init__(self, name, grade):
        super().__init__(name)    # Call Person's constructor
        self.grade = grade

    def show(self):
        print(f"Name: {self.name}, Grade: {self.grade}")


s = Student("krunal", "A")
s.show()
```

# BankAccount → SavingsAccount

```python
class BankAccount:
    def __init__(self, balance):
        self.balance = balance

    def account_info(self):
        print(f"Balance: Rs.{self.balance}")

class SavingsAccount(BankAccount):
    def account_info(self):
        super().account_info()
        print("Account type: Current")


sacc = SavingsAccount(1500000)
sacc.account_info()
```

# 🔎 SavingsAccount reuses and extends the account_info logic from BankAccount.

# 🔁 Method Overloading in Python

# Defining multiple methods in the same class with the same name, but different number or types of arguments.


# 🔴 BUT... Python does not support traditional method overloading
# 🔴 like Java or C++.

# 🔁 Instead, you can simulate it method overloading using:


# (1) Default arguments
# (2) *args (variable arguments)

# ✅ Simulating Method Overloading with Default Arguments

```python
class Calculator:
    def add(self, a=0, b=0, c=0):
        return a + b + c

calc = Calculator()
print(calc.add(5))         # 5 (5 + 0 + 0)
print(calc.add(5, 10))     # 15 (5 + 10 + 0)
print(calc.add(5, 10, 20)) # 35 (5 + 10 + 20)
```

# ✅ Simulating Method Overloading with *args

```python
class Calculator:
    def add(self, *numbers):
        return sum(numbers)

calc = Calculator()
print(calc.add(1, 2))         # 3
print(calc.add(1, 2, 3))      # 6
print(calc.add(1, 2, 3, 4))   # 10
```

# 🔁 Operator Overloading

# Operator overloading in Python allows you to define custom behavior for standard operators (+, -, *, etc.)
# when they're used with objects of your own classes.

# Instead of using operators in the default way, you can override their behavior by implementing dunder (magic) methods in your class.


# 💻 CODE

```python
class Phynix:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Phynix(self.x + other.x, self.y + other.y)

    def __repr__(self):
        return f"Phynix ({self.x}, {self.y})"

p1 = Phynix(2, 3)  #Self.x = 2     Self.y = 3
p2 = Phynix(4, 1)  #other.x = 4   other.y = 1

print(p1 + p2)
```

# 💻 CODE

```python
class ComplexNumber:
    def __init__(self, real, imag):
        self.real = real
        self.imag = imag


    def __add__(self, other):
        return ComplexNumber(self.real + other.real, self.imag + other.imag)


    def __eq__(self, other):
        return self.real == other.real and self.imag == other.imag


    def __repr__(self):
        return f"{self.real} + {self.imag}i"


# Creating complex number objects
c1 = ComplexNumber(3, 2)
c2 = ComplexNumber(1, 7)
c3 = ComplexNumber(3, 2)


# Using overloaded '+' and '==' operators
print("Addition:", c1 + c2)      # Output: 4 + 9i
print("Equality check:", c1 == c3)  # Output: True
```
