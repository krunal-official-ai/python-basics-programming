# 🔑 Core OOP Concepts in Python

# 🎯 1. Class
# ➵ A class is a blueprint for creating objects.

# 🎯 2. Object
# ➵ An object is an instance of a class.

# 🎯 3. Encapsulation
# ➵ Binding data (variables) and methods (functions) together in one unit (class). You can hide internal details using private variables.


# 🎯 4. Abstraction
# ➵ Hiding complex details and showing only the essential features.


# 🎯 5. Inheritance

# ➵ One class can inherit features from another. Promotes code reusability.


# 🎯 6. Polymorphism
# ➵ Same function or method behaves differently based on the object.


# 🧩 Class :
# A class is like a blueprint for creating objects.
# It defines a set of attributes and methods that the created objects (instances) will have.


# syntax :

```python
class ClassName:
    # constructor
    def __init__(self, attribute1, attribute2):
        self.attribute1 = attribute1
        self.attribute2 = attribute2

    # method
    def display(self):
        print(f"{self.attribute1}, {self.attribute2}")
```

# 🧩 Object
# An object is an instance of a class. It represents a specific realization of the class.

# syntax :

```python
# creating object of ClassName
obj = ClassName("value1", "value2")

# calling method
obj.display()
```

# 🚀 self Keyword
# ➵ self represents the instance of the class.

# ➵ It is used to access attributes and methods of the class in instance methods.

# ➵ It must be the first parameter of instance methods (including __init__), though it’s passed automatically when calling the method.

```python
class Student:
    def __init__(self, name):
        self.name = name  # setting instance attribute

    def show(self):
        print("Student name:", self.name)  # accessing attribute using self

s1 = Student("John")
s1.show()
```

# 🚀 Destructor : del Keyword

# ➵ del is used to delete objects, object attributes, or elements from a collection.

# ➵ When used on an object, it calls the destructor method __del__( ),
# ➵ which allows cleanup when an object is about to be destroyed.

# 💻 Deleting an object

```python
class Car:
    def __init__(self, brand):
        self.brand = brand

    def __del__(self):
        print(f"{self.brand} object is being deleted")

c1 = Car("Toyota")
del c1  # deletes the object and triggers __del__()
```

# 💻 Deleting an attribute

```python
class Car:
    def __init__(self, brand):
        self.brand = brand

c1 = Car("BMW")
del c1.brand  # deletes only the 'brand' attribute
c1.brand()
```

# Key Points :
# _ _del_ _ ( ) is automatically called when an object is garbage collected.

# Python uses automatic garbage collection, so __del__() may not be called immediately after del obj (especially in complex programs).

# You should not rely on destructors for critical cleanup (like closing files or network connections). Use with statements or context managers instead.
