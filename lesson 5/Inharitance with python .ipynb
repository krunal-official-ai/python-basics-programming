# ✅ Inheritance in Python is a feature of object-oriented programming

# ✅ Inheritance is a way to create a new class from an existing class.

# ➵ The new class is called the Child class (or subclass).

# ➵ The existing class is called the Parent class (or superclass).

# ➵ The child class inherits the methods and properties of the parent class.

# ➵ This helps to reuse code and create hierarchies in your program.

# ✅ Basic Syntax of Inharitance

```python
class Parent:
    # parent class code

class Child(Parent):
    # child class inherits from Parent
```

```python
class Parent:
    def Hi(self):
        print("Parent")

class Child(Parent):
    def say_hello(self):
        print("Bacchu")


# Example usage
N = Child()
N.Hi()       # Inherited from Parent
N.say_hello()   # Defined in Child
```

# 🧩 Types of Inheritance

# 🎯 Single Inheritance
# ➵ A child class inherits from one parent class.
# ➵ Single inheritance – One parent, one child

```python
class A:
    def show(self):
        print("A")

class B(A):
    pass
```

# Employee → Manager

```python
class Employee:
    def __init__(self, name):
        self.name = name

    def work(self):
        print(f"{self.name} is working.")

class Manager(Employee):  # Inherits from Employee
    def manage(self):
        print(f"{self.name} is managing the Ai Task force.")


m = Manager("Krunal")
m.work()    # inherited from Employee
m.manage()  # defined in Manager

```

# Vehicle → Car

```python
class Vehicle:
    def __init__(self, brand):
        self.brand = brand

    def move(self):
        print(f"{self.brand} is moving.")

class Car(Vehicle):  # Inherits from Vehicle
    def play_radio(self):
        print(f"{self.brand}'s radio is playing.")

c = Car("Toyota")
c.move()         # inherited
c.play_radio()   # specific to Car
```

# BankAccount → SavingsAccount

```python
class BankAccount:
    def __init__(self, balance):
        self.balance = balance

    def check_balance(self):
        print(f"Balance: Rs.{self.balance}")

class SavingsAccount(BankAccount):  # Single inheritance
    def add_interest(self):
        interest = self.balance * 0.05
        self.balance += interest
        print(f"Interest added: Rs.{interest}")


sa = SavingsAccount(1000)
sa.check_balance()   # inherited
sa.add_interest()    # specific to SavingsAccount
```

# Teacher → Student

```python
class Person:
    def __init__(self, name):
        self.name = name

    def introduce(self):
        print(f"I am {self.name}.")

class Student(Person):
    def study(self):
        print(f"{self.name} is studying.")


s = Student("Karma")
s.introduce()  # inherited
s.study()      # specific
```

# 🎯 Multiple Inheritance
# ➵ A child class inherits from multiple parent classes.
# ➵ Multiple inheritance – Child inherits from more than one parent

```python
class A:
    def method_a(self):
        print("A")

class B:
    def method_b(self):
        print("B")

class C(A, B):
    pass
```

# Person + Programmer → Employee

```python
class Person:
    def __init__(self, name):
        self.name = name

    def speak(self):
        print(f"{self.name} is speaking.")

class Programmer:
    def code(self):
        print(f"{self.name} is writing code.")


class Employee(Person, Programmer):  # Inherits from both
    def work(self):
        print(f"{self.name} is working in the company.")



e = Employee("Krunal")
e.speak()     # from Person
e.code()      # from Programmer
e.work()      # from Employee
```

# OnlineOrder + Payment → ECommerce

```python
class OnlineOrder:
    def place_order(self):
        print("Order placed online.")

class Payment:
    def process_payment(self):
        print("Payment processed.")

class ECommerce(OnlineOrder, Payment):  # Multiple inheritance
    def track_order(self):
        print("Order is being tracked.")


www = ECommerce()
www.place_order()       # from OnlineOrder
www.process_payment()   # from Payment
www.track_order()       # from ECommerce
```

# 🎯 Multilevel Inheritance
# ➵ A class inherits from a child class which in turn inherits from another class.
# ➵ Multilevel inheritance – Child → Parent → Grandparent

```python
class A:
    def spark_a(self):
        print("A")

class B(A):
    def spark_b(self):
        print("B")

class C(B):
    def spark_c(self):
        print("C")
```

# Person → Employee → Manager

```python
class Person:
    def __init__(self, name):
        self.name = name

    def introduce(self):
        print(f"I am {self.name}.")

class Employee(Person):
    def work(self):
        print(f"{self.name} is working.")

class Manager(Employee):
    def manage_team(self):
        print(f"{self.name} is managing the Ai Developers")


m = Manager("Krunal")
m.introduce()     # from Person
m.work()          # from Employee
m.manage_team()   # from Manager
```

# Vehicle → Car → ElectricCar

```python
class Vehicle:
    def move(self):
        print("Vehicle is moving.")

class Car(Vehicle):
    def play_radio(self):
        print("Car radio is playing.")

class ElectricCar(Car):
    def charge_battery(self):
        print("Charging electric battery.")


tesla = ElectricCar()
tesla.move()           # from Vehicle
tesla.play_radio()     # from Car
tesla.charge_battery()  # from ElectricCar
```

# Vehicle System

```python
class Vehicle:  # Parent class
    def __init__(self, brand):
        self.brand = brand

    def start_engine(self):
        print(f"{self.brand} engine started.")

class Car(Vehicle):  # Child class 1
    def open_trunk(self):
        print("Trunk is now open.")

class Bike(Vehicle):  # Child class 2
    def kick_start(self):
        print("Bike kick-started manually.")



car1 = Car("Toyota")
car1.start_engine()     # Inherited
car1.open_trunk()       # Specific to Car

bike1 = Bike("Yamaha")
bike1.start_engine()    # Inherited
bike1.kick_start()      # Specific to Bike

```

# School System

```python
class Principal:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def introduce(self):
        print(f"My name is {self.name} and I'm {self.age} years old.")

class Student(Principal):
    def study(self):
        print(f"{self.name} is studying.")

class Teacher(Principal):
    def teach(self):
        print(f"{self.name} is teaching.")



s = Student("Karma", 2)
s.introduce()  # Inherited
s.study()      # Specific to Student

t = Teacher("Mr.Krunal", 30)
t.introduce()  # Inherited
t.teach()      # Specific to Teacher

```

# 🎯 Hybrid Inheritance
# ➵ A combination of two or more types of inheritance.
# ➵ Hybrid inheritance – Combination of above types
# ➵ Single Inheritance | Multiple Inheritance
# ➵ Multilevel Inheritance | Hierarchical Inheritance

```python
class A:
    def method_a(self):
        print("A")

class B(A):
    def method_b(self):
        print("B")

class C(A):
    def method_c(self):
        print("C")

class D(B, C):
    def method_d(self):
        print("D")

d = D()
d.method_a()  # From class A
```

# Educational System

       Principal
         |
    -------------
    |           |
 Student     Teacher
    |
 ResearchStudent
         |
   TeachingAssistant


```python
class Pricipal:
    def __init__(self, name):
        self.name = name

    def introduce(self):
        print(f"I am {self.name}.")

class Student(Pricipal):  # Hierarchical from Person
    def study(self):
        print(f"{self.name} is studying.")

class Teacher(Pricipal):  # Hierarchical from Person
    def teach(self):
        print(f"{self.name} is teaching.")

class ResearchStudent(Student):  # Multilevel inheritance
    def research(self):
        print(f"{self.name} is doing research.")

class TeachingAssistant(Teacher, ResearchStudent):  # Multiple inheritance
    def assist(self):
        print(f"{self.name} is assisting in teaching.")


ES = TeachingAssistant("Krunal")

ES.introduce()    # from Person
ES.teach()        # from Teacher
ES.study()        # from Student
ES.research()     # from ResearchStudent
ES.assist()       # from TeachingAssistant
```
