# 🧩 Built-in dunder methods

# dunder methods

# short for "double underscore" methods,
# also known as magic methods
# and special methods
# this method have built-in methods that have double underscores before and after their names — like __init__, __str__, __len__, etc.



# They enable you to define how objects of your classes behave with built-in functions, operators, and syntactic features.

# 🚀 Common Built-in Dunder Methods


# 📦 Object Lifecycle

```python
|      Method           |            Purpose            |
| --------------------- | ----------------------------- |
|   __init__(self, ...) | Constructor (initializer)     |
|  __new__(cls, ...)    | Low-level object creation     |
|  __del__(self)        | Destructor (called on delete) |
```

# 🧾 String & Representation

```python
| Method           | Used For                           |
| ---------------- | ---------------------------------- |
|  __str__(self)   | String representation (`str()`)    |
|  __repr__(self)  | Official representation (`repr()`) |
|  __format__()    | Custom formatting (`format()`)     |
```

# 🔢 Numeric Operators

```python
| Method           | Operator  |
| ---------------- | --------- |
|  __add__()       | `+`       |
|  __sub__()       | `-`       |
|  __mul__()       | `*`       |
|  __truediv__()   | `/`       |
|  __floordiv__()  | `//`      |
|  __mod__()       | `%`       |
|  __pow__()       | `**`      |
|  __neg__()       | Unary `-` |
```

# 📏 Length, Containment, Iteration

```python
| Method           | Used For                      |
| ---------------- | ----------------------------- |
|  __len__()       | Length (`len()`)              |
|  __contains__()  | Containment (`in`)            |
|  __getitem__()   | Indexing/slicing (`obj[i]`)   |
|  __setitem__()   | Setting values (`obj[i] = v`) |
|  __delitem__()   | Deleting items (`del obj[i]`) |
|  __iter__()      | Iteration (`for x in obj`)    |
|  __next__()      | Next value (`next(obj)`)      |
```

# ⚖️ Comparisons

```python
| Method     | Operator |
| ---------- | -------- |
|  __eq__()  | `==`     |
|  __ne__()  | `!=`     |
|  __lt__()  | `<`      |
|  __le__()  | `<=`     |
|  __gt__()  | `>`      |
|  __ge__()  | `>=`     |
```

# 📞 Callable & Context

```python
| Method        | Used For                       |
| ------------- | ------------------------------ |
| __call__()    | Making objects callable        |
| __enter__()   | Context manager entry (`with`) |
| __exit__()    | Context manager exit (`with`)  |
```

# 🧾 Miscellaneous

```python
| Method           | Purpose                   |
| ---------------- | ------------------------- |
|  __bool__()      | Truth value (`bool(obj)`) |
|  __hash__()      | Hashing (`hash(obj)`)     |
|  __copy__()      | Shallow copy              |
|  __deepcopy__()  | Deep copy                 |
```
