# 🧩 Reflection-Enabling Functions

# Reflection in Python means the ability of a program to inspect and manipulate its own structure and behavior at runtime.

# 🧩 Common Reflection-Enabling Functions

# |Function|----------------------------|Description|

# 1. type( ) -------------------> Returns the type of an object
# 2. id( )	----------------------> Returns the memory address of an object
# 3. dir( ) ---------------------> Lists the attributes and methods of an object
# 4. hasattr( ) ----------------> Checks if an object has a given attribute
# 5. getattr( ) -----------------> Gets the value of an attribute
# 6. setattr( )	-----------------> Sets the value of an attribute
# 7. delattr( ) -----------------> Deletes an attribute
# 8. isinstance( ) -------------> Checks if an object is an instance of a class
# 9. issubclass( ) ------------>  Checks if a class is a subclass of another class

# 💻 CODE

```python
class Person:
    name = "Krunal"

p = Person()

print(type(p))          # <class '__main__.Person'>
print("         ")
print(dir(p))
print("         ")

# Lists all methods/attributes of 'p'
print(hasattr(p, 'name'))  # True

print(getattr(p, 'name'))  # Krunal

setattr(p, 'age', 30)      

print(p.age)               # 30

delattr(p, 'name')         
```
