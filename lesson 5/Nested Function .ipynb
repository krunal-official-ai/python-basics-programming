# 🔁 Nested Function in Python


# A nested function (also called an inner function) is a function defined inside another function.

```python
def outer_function():
    def inner_function():
        print("This is the inner function")

    print("This is the outer function")
    inner_function()  # Call inner function
```

# 💻 CODE

```python
def OUTER(name):
    def INNER():
        return "Welcome to Our A.I WORLD!"

    print(f"Hello {name}, {INNER()}")

OUTER("KRUNAL")
```

```python
def outer(x):
    def inner(y):
        return x + y
    return inner

add_five = outer(5)
print(add_five(3))  # Output: 8
```

# Here, inner( ) remembers x = 5 even after outer( ) is done — that’s called a closure.
