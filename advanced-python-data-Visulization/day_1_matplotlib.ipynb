```python
import matplotlib.pyplot as plt
```

##### Bar Chart

```python
x = ['Math', 'Science', 'English']
y = [85, 90, 70]
```

```python
plt.bar(x, y)

plt.title("Subject-wise Marks")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show()
```

```python
plt.bar(x, y, width=0.5)  # width must be in range of 0.0 to 1.0

plt.title("Subject-wise Marks")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show()
```

```python
plt.bar(x, y, color='green')

plt.title("Subject-wise Marks")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show()
```

```python
plt.bar(x, y, color=['red', 'green', 'blue'])

plt.title("Subject-wise Marks")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show()
```

```python
plt.bar(x, y, alpha=0.5)  # alpha is opacity, ranges from 0.0 to 1.0

plt.title("Subject-wise Marks")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show()
```

```python
plt.figure(figsize=(8, 4))  # figsize is in inches

plt.bar(x, y, align='center')  # align can be 'center' or 'edge'

plt.title("Subject-wise Marks")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show()
```

```python
plt.bar(x, y, color='tomato', edgecolor='black', linewidth=3)  # edge color, edge width

plt.title("Subject-wise Marks")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show()
```

```python
plt.bar(x, y, color='tomato', edgecolor='black', linewidth=2)  # edge color, edge width

plt.title("Subject-wise Marks")
plt.xlabel("Subjects")
plt.ylabel("Marks")

# Optional: add value labels on top of bars

# plt.text(0, 85 + 1, str(85), ha='center', fontweight='bold')
# plt.text(1, 90 + 1, str(90), ha='center', fontweight='bold')
# plt.text(2, 70 + 1, str(70), ha='center', fontweight='bold')

for i, v in enumerate(y):
    plt.text(i, v+1, str(v), ha='center', fontweight='bold')

# plt.tight_layout()  # adjust layout to prevent clipping of tick-labels
plt.show()
```

```python
# Horizontal Bar Chart

plt.barh(x, y, color='purple', edgecolor='black', linewidth=2)  # edge color, edge width

plt.title("Subject-wise Marks")
plt.xlabel("Subjects")
plt.ylabel("Marks")

# Optional: add value labels on top of bars

for i, v in enumerate(y):
    plt.text(v + 1, i, str(v), va='center', fontweight='bold')

plt.show()
```

##### Line Chart

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
revenue = [5000, 5200, 5800, 6200, 6000, 6700, 7100, 6900, 7200, 7500, 7700, 8000]
```

```python
plt.plot(months, revenue)       # use this if using a dataframe => plt.plot(df['months'], df['revenue']) 

plt.title("Monthly Revenue")
plt.xlabel("Months")
plt.ylabel("Revenue")
plt.show()
```

```python
plt.plot(months, revenue, marker="o")
plt.show()
```

```python
plt.plot(months, revenue, marker="+")
plt.show()
```

```python
plt.plot(months, revenue, marker="*")
plt.show()
```

```python
plt.plot(months, revenue, marker="o", markersize=8, color="black", markerfacecolor="red")
plt.show()
```

```python
plt.figure(figsize=(12, 8))

plt.plot(months, revenue, marker="o", markersize=8, color="black", markerfacecolor="red", linewidth=2)

for i, val in enumerate(revenue):
    plt.text(i, val + 120, str(val), ha='center', fontweight='bold')

plt.grid(True, linestyle="--", alpha=0.7, axis='both')

plt.savefig('monthly_revenue.png', dpi=300)

plt.show()
```

##### Scatter Chart

```python
ad_spend = [5, 6, 7, 6, 8, 7.5, 9, 10, 9.5, 11, 10.5, 12]
sales = [40, 45, 50, 46, 54, 53, 58, 62, 59, 65, 64, 70]
months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
```

```python
plt.scatter(ad_spend, sales)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(ad_spend, sales, color='orange', edgecolors='black', s=100)  # s is size of the marker
plt.show()
```

```python
plt.figure(figsize=(10, 8))

plt.scatter(ad_spend, sales, color='orange', edgecolors='black', s=100)  # s is size of the marker

plt.xlabel('Ad Spend')
plt.ylabel('Sales')

for i, month in enumerate(months):
    plt.text(ad_spend[i], sales[i] + 1.5, months[i], ha='center', fontweight='bold')

plt.show()
```

##### Pie Chart

```python
products = ['Laptops', 'Smartphones', 'Tablets']
sales = [8, 12, 3]
```

```python
plt.pie(sales)
plt.show()
```

```python
plt.pie(sales, startangle=90)
plt.show()
```

```python
plt.pie(sales, startangle=90, labels=products)
plt.show()
```

```python
plt.pie(sales, startangle=90, labels=products, autopct='%.2f%%')
plt.show()
```

```python
plt.pie(sales, startangle=90, labels=products, autopct='%.2f%%', colors=['tomato', 'pink', 'lightblue'])
plt.show()
```

```python
plt.pie(sales, startangle=90, labels=products, autopct='%.2f%%', explode=(0.0, 0.0, 0.08))
plt.show()
```

```python
plt.pie(
    sales, 
    startangle=90, 
    labels=products, 
    autopct='%.2f%%', 
    explode=(0.02, 0.05, 0.08),
    shadow=True, 
    textprops={'fontsize': 14, 'color': 'black'},
    )
plt.show()
```

```python
plt.pie(sales, startangle=90, labels=products, autopct='%.2f%%', colors=['tomato', 'pink', 'lightblue'])

centre_circle = plt.Circle((0,0), 0.4, color='white')
plt.gcf().gca().add_artist(centre_circle)

plt.show()
```

```python
plt.pie(
    sales, 
    startangle=90, 
    labels=products, 
    autopct=lambda pt: f'{pt:.2f}%' if pt>20 else '', 
    colors=['tomato', 'pink', 'lightblue']
    )

plt.legend(loc='best')
plt.text(0, 0, 'Sales', ha='center', va='center', fontsize=16)
plt.show()
```

##### Histogram

```python
# Sample Data: Math exam scores
scores = [45, 78, 56, 89, 90, 67, 80, 56, 92, 45,
          60, 75, 85, 55, 70, 90, 88, 72, 60, 65,
          68, 77, 85, 95, 48, 52, 60, 75, 80, 82]
```

```python
# plt.hist(scores)

plt.hist(scores, bins=10)
```

```python
plt.hist(scores, bins=[45, 55, 65, 75, 85, 95], edgecolor='black')
```

```python
# Compare two scores (Maths & Science)
math_scores =    [45, 78, 56, 89, 90, 67, 80, 56, 92, 45,
                  60, 75, 85, 55, 70, 90, 88, 72, 60, 65,
                  68, 77, 85, 95, 48, 52, 60, 75, 80, 82]

science_scores = [40, 70, 60, 85, 88, 66, 75, 50, 90, 42,
                  58, 72, 80, 50, 65, 88, 86, 70, 55, 60,
                  63, 74, 83, 91, 45, 51, 58, 73, 78, 79]
```

```python
plt.hist(math_scores, bins=5, alpha=0.3, label='Maths', color='tomato', edgecolor='black')
plt.hist(science_scores, bins=5, alpha=0.3, label='Science', color='skyblue', edgecolor='black')

plt.legend(loc='best')
plt.title('Comparison of Math and Science Scores')
plt.show()
```
