```python
import matplotlib.pyplot as plt
```

##### Stack Chart

```python
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri']
product_A = [5, 7, 6, 4, 5]
product_B = [3, 5, 2, 6, 4]
product_C = [2, 3, 4, 3, 2]
```

```python
plt.bar(days, product_A, label='Product A')
plt.bar(days, product_B, label='Product B')
plt.bar(days, product_C, label='Product C')

plt.legend()

plt.show()
```

```python
# Research this cell

plt.bar(days, product_A, label='Product A')
plt.bar(days, product_B, label='Product B', bottom=product_A) 

position_c = [i1+i2 for i1, i2 in zip(product_A, product_B)]

plt.bar(days, product_C, label='Product C', bottom=position_c) 

plt.legend(loc="best")
plt.grid(axis='y', alpha=0.4)

plt.show()
```

```python
plt.bar(days, product_A, label='Product A', width=0.6)
plt.bar(days, product_B, label='Product B', width=0.4)
plt.bar(days, product_C, label='Product C', width=0.2)

plt.legend(loc="best")

plt.show()
```

```python
# Horizontal Stacked Bar Chart
categories = ['A', 'B', 'C']
group1 = [4, 6, 3]
group2 = [5, 2, 7]

plt.barh(categories, group1, label='Group 1')
plt.barh(categories, group2, left=group1, label='Group 2')

plt.title('Horizontal Stacked Bar')
plt.xlabel('Values')
plt.ylabel('Categories')
plt.legend()
plt.show()
```

```python
# Stacked Area Chart (Multiple Line Plots with Fill => Stacked Area Chart)

plt.stackplot(
    days, 
    product_A, 
    product_B, 
    product_C, 
    labels=['Product A', 'Product B', 'Product C']
    )

plt.legend()
plt.grid(axis='y', alpha=0.4)

plt.show()
```

##### Subplots

```python
# Syntax
# ------
# fig, axs = plt.subplots(nrows, ncols)
```

```python
x = [1, 2, 3, 4, 5]
y1 = [10, 20, 30, 40, 50]
y2 = [5, 15, 25, 35, 45]

fig, axs = plt.subplots(1, 2, figsize=(10, 4))

axs[0].plot(x, y1, color='blue')
axs[0].set_title("Line Plot")

axs[1].bar(x, y2, color='green')
axs[1].set_title("Bar Chart")

plt.tight_layout()
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

fig, axs = plt.subplots(2, 2, figsize=(10, 6))

axs[0, 0].plot(x, y)
axs[0, 0].set_title("Plot 1")

axs[0, 1].bar(x, y)
axs[0, 1].set_title("Bar 1")

axs[1, 0].scatter(x, y)
axs[1, 0].set_title("Scatter 1")

axs[1, 1].barh(x, y)
axs[1, 1].set_title("Horizontal Bar")

plt.tight_layout()
plt.show()
```

##### Legends

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y1 = [1, 4, 2, 3]
y2 = [2, 3, 4, 1]

plt.plot(x, y1, label='Line 1')
plt.plot(x, y2, label='Line 2')

plt.title('Legend Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.legend()  # Show legend
# plt.legend(loc='upper left')       # Position
# plt.legend(fontsize=12)             # Font size
# plt.legend(title='Lines')           # Title for the legend box
# plt.legend(frameon=False)           # Remove box/frame
plt.show()
```

```python
# Place Legend Outside the Plot

x = [1, 2, 3, 4]
y1 = [1, 4, 2, 3]
y2 = [2, 3, 4, 1]

plt.plot(x, y1, label='Line A')
plt.plot(x, y2, label='Line B')

plt.title('Legend Outside the Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Place legend outside
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.tight_layout()  # Adjust plot to not clip the legend

plt.show()
```

```python
# Add Styling: Shadow, Title, Frame, Font Size

plt.plot(x, y1, label='Line A')
plt.plot(x, y2, label='Line B')

plt.title('Legend Outside the Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.legend(
    loc='upper left',
    title='Legend Title',
    fontsize='medium',
    title_fontsize='large',
    shadow=True,
    frameon=True,
    facecolor='lightyellow',
    edgecolor='black'
)

plt.tight_layout() 

plt.show()
```

```python
# Multiple Legends
line1, = plt.plot(x, y1, label='Line A')
line2, = plt.plot(x, y2, label='Line B')

first_legend = plt.legend(handles=[line1], loc='upper left')
plt.gca().add_artist(first_legend)              # Add the first legend manually
plt.legend(handles=[line2], loc='lower right')  # Second legend
```
