### **Figure-level vs. axes-level functions**

 There is a cross-cutting classification of seaborn functions as “**axes-level**” or “**figure-level**”

 1. **Figure-level functions:** relplot, displot, catplot
 2. **Axes-level functions:** scatterplot, lineplot, histplot, kdeplot, ecdfplot, rugplot, stripplot, swarmplot, boxplot, violinplot, pointplot, barplot

## 1. Bar Plot - Titanic Dataset

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load dataset
df = sns.load_dataset("titanic")

# Bar plot
plt.figure(figsize=(8, 5))
sns.barplot(x="class", y="fare", data=df, estimator=np.sum, errorbar=None)
plt.xlabel("Passenger Class")
plt.ylabel("Total Fare")
plt.title("Total Fare by Class")
plt.show()
```

```python
# Real-world styled dataset
df = pd.DataFrame({
    'Category': ['Laptop', 'Laptop', 'Laptop', 'Phone', 'Phone', 'Tablet', 'Tablet', 'Tablet'],
    'Rating': [4.5, 4.0, 4.2, 4.8, 4.6, 4.1, 4.0, 3.9]
})

# Bar plot
sns.barplot(x='Category', y='Rating', data=df, errorbar=None, palette='deep')
plt.title('Average Customer Ratings by Product Category')
plt.ylabel('Avg Rating')
plt.xlabel('Product Category')
plt.show()
```

```python
df = pd.DataFrame({
    'Category': ['Laptop', 'Laptop', 'Phone', 'Phone', 'Tablet', 'Tablet'],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female'],
    'Rating': [4.5, 4.1, 4.7, 4.6, 4.0, 4.1]
})

sns.barplot(x='Category', y='Rating', data=df, palette='Set2', errorbar=None, hue='Gender')
plt.title('Avg Ratings by Product & Gender')
plt.show()
```

## 2. Line Plot - Flights Dataset

```python
flights = sns.load_dataset("flights")
line_data = flights.groupby("year")["passengers"].sum().reset_index()

plt.figure(figsize=(10, 5))
sns.lineplot(data=line_data, x="year", y="passengers", marker="o")
plt.title("Total Passengers per Year")
plt.show()
```

## 3. Scatter Plot with Regression Line - Tips Dataset

```python
tips = sns.load_dataset("tips")

plt.figure(figsize=(8, 5))
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time")
# sns.regplot(data=tips, x="total_bill", y="tip", scatter=False, color='black')
plt.title("Total Bill vs Tip with Regression Line")
plt.show()
```

## 4. Pie Chart - Titanic Dataset

```python
# Pie chart can't be created directly with seaborn, but we can use matplotlib
# to create a pie chart for the survival rate
# on the Titanic dataset

df = sns.load_dataset("titanic")

survival_counts = df['survived'].value_counts()
labels = ['Did Not Survive', 'Survived']
colors = ['lightcoral', 'lightskyblue']

plt.figure(figsize=(6, 6))
plt.pie(survival_counts, labels=labels, autopct='%.2f%%', colors=colors)
plt.title("Survival Rate on Titanic")
plt.show()
```

## 5. Histogram - Titanic Dataset

```python
plt.figure(figsize=(8, 5))
sns.histplot(df["age"].dropna(), bins=30, kde=True)
plt.title("Age Distribution on Titanic")
plt.show()
```

## 6. Count Plot - Titanic Dataset

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="class", hue="sex")
plt.title("Passenger Count by Class and Gender")
plt.show()
```

## 7. Heatmap - Correlation in Titanic Dataset

```python
corr = df.corr(numeric_only=True)
plt.figure(figsize=(10, 6))
sns.heatmap(corr, annot=True, linewidths=0.5, cmap='YlOrRd')
plt.title("Correlation Heatmap - Titanic")
plt.show()
```
